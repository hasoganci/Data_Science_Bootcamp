1. Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3$\pi$ and 3$\pi$ with 0.1 intervals.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(10, 5))

# Data for plotting
t = np.arange(-3*np.pi, 3*np.pi, 0.1)
s = np.cos(t)
plt.plot(t,s)
plt.xlabel('time (s)')
plt.ylabel('cos value')
plt.title('cosine function for the numbers beetwen -3 pi  and 3 pi')
plt.show()
```

2. This time, create the same plot in red color with object methods. 

```python
plt.figure(figsize=(12, 6))
fig, ax = plt.subplots()
ax.plot(t,s, c = "red")
plt.show()
```
